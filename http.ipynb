```python
import os
import json
import requests
```

```python
# 设置模型，默认就是bge-m3，之后会支持更多模型
# response = requests.post("http://127.0.0.1:8000/set-model/", json={"model_name": "bge-m3"})
```

```python
# 索引

base_dir = "/data/peitian/Data/legal/yidaiyilu/output"
texts = []
for path in os.listdir(base_dir):
    path = os.path.join(base_dir, path)
    with open(path) as f:
        for i, line in enumerate(f):
            item = json.loads(line)
            texts.append(item["text"])

response = requests.post("http://127.0.0.1:8000/index", json={
    # 要索引的文本
    "texts": texts,
    # 索引名（该名称下会索引上述文本，如果索引已经存在，则会append）
    "index_name": "test",
})
response.text
```

```python
# 搜索

response = requests.post("http://127.0.0.1:8000/search", json={
    # 查询（应该可以async查）
    "query": "反垄断",
    # 索引名
    "index_name": "test"
})
json.loads(response.text)
```

```python
# 删除索引

response = requests.post("http://127.0.0.1:8000/remove-index/", json={"index_name": "test"})
response.text
```
